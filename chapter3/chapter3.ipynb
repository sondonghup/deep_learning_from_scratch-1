### 3.2.4 시그모이드 함수 구현하기

````python
import numpy as np
import matplotlib.pylab as plt

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

num_list = np.array([1.0, -1.0, 2.0, 100.0])

print(sigmoid_function(num_list))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
````

### 3.2.5 시그모이드 함수와 계단 함수 비교

````python
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    y = x > 0
    return y.astype(np.int)

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

x0 = np.arange(-5.0, 5.0, 0.1)
x1 = np.arange(-5.0, 5.0, 0.1)
y0 = step_function(x0)
y1 = sigmoid_function(x1)
plt.plot(x0, y0, x1, y1)
plt.ylim(-0.1, 1.1)
plt.xlim(-5.1, 5.1)
plt.show()
````

##### 두함수는 모두 0과 1사이를 나타내지만 계단 함수 같은 경우 극단적이여서 연속적인 실수를 범할 수 가 있습니다.

### 3.2.6 비선형 함수

##### 앞에서 말한 두 계단 함수와 시그모이드 함수는 비선형 함수입니다.
##### 선형 함수 : 직선으로 표현이 가능한 함수
##### 비선형 함수 : 1개의 직선으로 표현이 불가능한 함수
##### 신경망에서는 선형함수가 아닌 비선형 함수를 사용해야 합니다. 선형함수는 은닉층을 안 거치고 표현을 할 수 있어 층을 쌓는 장점을 발휘할 수 가 없습니다.

##### -> 예를 들어 h(x) = 3x 라는 층을 세개 쌓는다고 생각을 해봅시다. 그럼 a = 3x, b = 3a, y = 3b로 표현이 되고 이를 간단히 하면 y = 3 * 3 * 3x 가 됩니다. 그렇다면 아에 h(x) = 27x 라는 한개의 식으로 표현을 할 수 있어 은닉층없이 표현을 할 수 있게 됩니다.

### 3.2.7 ReLU 함수

##### ReLU함수는 입력이 0을 넘을땐 값을 그대로 출력하나 0 이하일땐 0으로 출력합니다.

````python
import numpy as np
import matplotlib.pylab as plt

def relu_function(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 2.1)
plt.show()
````
